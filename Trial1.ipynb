```python
import numpy as np

# Create a 2x3 array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# Print the array
print("Original Array:")
print(arr)

# Perform some basic operations
print("\nOperations:")
print("Sum of all elements:", np.sum(arr))
print("Mean of all elements:", np.mean(arr))
print("Transpose of the array:")
print(np.transpose(arr))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate some sample data
num_points = 100
x = np.linspace(0, 10, num_points)
y = np.sin(x) + np.random.normal(0, 0.1, num_points)

# Create a Pandas DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Plot the data using Matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(data['x'], data['y'], label='Data Points')
plt.title('Sample Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

# Fit a linear trendline using NumPy
coefficients = np.polyfit(data['x'], data['y'], 1)
trendline = np.polyval(coefficients, data['x'])

# Plot the trendline
plt.plot(data['x'], trendline, color='red', label='Trendline')

# Show the plot
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Generate random data
np.random.seed(0)
num_points = 50
start_date = datetime(2022, 1, 1)
dates = [start_date + timedelta(days=i) for i in range(num_points)]
values = np.random.randint(50, 100, size=num_points)

# Create a Pandas DataFrame
data = pd.DataFrame({'Date': dates, 'Value': values})

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(data.head())

# Perform some basic operations with NumPy
mean_value = np.mean(data['Value'])
median_value = np.median(data['Value'])
max_value = np.max(data['Value'])
min_value = np.min(data['Value'])

print("\nStatistics:")
print("Mean Value:", mean_value)
print("Median Value:", median_value)
print("Max Value:", max_value)
print("Min Value:", min_value)

# Plot the data using Matplotlib
plt.figure(figsize=(8, 6))
plt.plot(data['Date'], data['Value'], marker='o', linestyle='-', color='b', label='Data')
plt.title('Sample Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Highlight points where the value exceeds the mean
plt.scatter(data[data['Value'] > mean_value]['Date'], data[data['Value'] > mean_value]['Value'], color='r', label='Above Mean')

# Show the plot
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
num_samples = 100
x = np.random.normal(loc=0, scale=1, size=num_samples)
y = 2 * x + np.random.normal(loc=0, scale=0.5, size=num_samples)

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Compute statistics
mean_x = np.mean(data['X'])
mean_y = np.mean(data['Y'])
covariance = np.cov(data['X'], data['Y'])[0, 1]
correlation = np.corrcoef(data['X'], data['Y'])[0, 1]

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['Y'], color='blue', alpha=0.5)
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add mean line
plt.axhline(mean_y, color='red', linestyle='--', label=f'Mean Y: {mean_y:.2f}')
plt.axvline(mean_x, color='green', linestyle='--', label=f'Mean X: {mean_x:.2f}')

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

# Display statistics
print("Statistics:")
print(f"Mean of X: {mean_x:.2f}")
print(f"Mean of Y: {mean_y:.2f}")
print(f"Covariance between X and Y: {covariance:.2f}")
print(f"Correlation between X and Y: {correlation:.2f}")
```
